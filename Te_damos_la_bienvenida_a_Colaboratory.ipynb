## Ciencia de datos
</div>

Con Colab, puedes aprovechar toda la potencia de las bibliotecas más populares de Python para analizar y visualizar datos. La celda de código de abajo utiliza <strong>NumPy</strong> para generar datos aleatorios y <strong>Matplotlib</strong> para visualizarlos. Para editar el código, solo tienes que hacer clic en la celda.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# Paso 1: Importar las bibliotecas necesarias
# -------------------------------------------
# Estas son las bibliotecas básicas que utilizaremos.
# - `load_iris` para cargar el conjunto de datos Iris.
# - `KMeans` es el algoritmo de agrupamiento que aplicaremos.
# - `PCA` nos ayudará a reducir las dimensiones para visualizar los datos.
# - `matplotlib.pyplot` se utiliza para generar gráficos.

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd

# Paso 2: Cargar el conjunto de datos Iris
# ----------------------------------------
# El conjunto de datos Iris viene incluido en Scikit-Learn.
# Contiene 150 muestras y 4 características (longitudes y anchuras de sépalos y pétalos)
# que describen tres especies de flores de iris.

iris = load_iris()

# Convertimos los datos en un DataFrame para manejarlos con mayor facilidad
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target  # Añadimos la columna de las etiquetas para referencia

# Paso 3: Seleccionar una muestra aleatoria del conjunto de datos
# ---------------------------------------------------------------
# Vamos a seleccionar una muestra del 50% del conjunto de datos para realizar
# un análisis más rápido. Utilizamos la función 'sample' de pandas.
# `frac=0.5` indica que seleccionamos el 50% de los datos.

df_sample = df.sample(frac=0.5, random_state=42)

# Separar las características (X) y la columna de etiquetas (y)
X_sample = df_sample.drop('target', axis=1)  # Características para clustering
y_sample = df_sample['target']  # Etiquetas reales (no usadas en K-Means, pero útiles para evaluación)

# Paso 4: Aplicar K-Means al subconjunto seleccionado
# ----------------------------------------------------
# Inicializamos el modelo K-Means con 3 clústeres, ya que sabemos que en el conjunto de datos Iris
# hay tres especies de flores.
# Ajustamos el modelo con el subconjunto de datos que seleccionamos (X_sample).

kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X_sample)

# Predecir los clústeres asignados por K-Means
labels = kmeans.predict(X_sample)

# Paso 5: Visualizar los resultados usando PCA para reducir las dimensiones
# -------------------------------------------------------------------------
# Como tenemos 4 características (dimensiones), usamos PCA para reducirlas a 2D,
# lo que nos permite visualizar los clústeres en un gráfico de dispersión.

pca = PCA(2)  # Reducción de las dimensiones a 2 componentes principales
X_pca = pca.fit_transform(X_sample)

# Graficar los puntos agrupados por K-Means
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, s=50, cmap='viridis')  # Puntos del subconjunto

# Graficar los centroides de los clústeres
centroids = pca.transform(kmeans.cluster_centers_)  # Transformar los centroides a 2D con PCA
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75, marker='X')  # Centroides

# Título del gráfico
plt.title('Clustering con K-Means en una muestra del conjunto de datos Iris')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```
